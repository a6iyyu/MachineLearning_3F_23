```python
import os
import pandas as pd
from google.colab import files
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
```

```python
data_path = "voice.csv"

if not os.path.exists(data_path):
    print(f"File '{data_path}' tidak ditemukan. Silakan upload file.")
    uploaded = files.upload()
    if data_path not in uploaded:
        raise FileNotFoundError(f"Upload dibatalkan atau nama file salah. '{data_path}' tidak ditemukan.")
    print(f"\nFile '{data_path}' berhasil di-upload.")
else:
    # Jika file sudah ada, lewati upload
    print(f"File '{data_path}' sudah ada. Proses upload dilewati.")
```

```python
df_voice = pd.read_csv(data_path)

le = LabelEncoder()
df_voice['label'] = le.fit_transform(df_voice['label'])

X = df_voice.drop('label', axis=1)
y = df_voice['label']

# SVM sangat sensitif terhadap skala data, jadi kita lakukan scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
split_ratios = [0.3, 0.2] # 70:30 (test_size=0.3) dan 80:20 (test_size=0.2)
kernels = ['linear', 'poly', 'rbf']

results = []

print("Memulai pelatihan model SVM untuk data voice.csv...")
```

```python
for split in split_ratios:
    split_label = f"{int((1-split)*100)}:{int(split*100)}"
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=split, random_state=42)

    for kernel_name in kernels:
        model = SVC(kernel=kernel_name, random_state=42)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        results.append({ "Split Ratio": split_label, "Kernel": kernel_name, "Accuracy": accuracy })
```

```python
results_df = pd.DataFrame(results)
results_df_sorted = results_df.sort_values(by="Accuracy", ascending=False)

print("\n--- Hasil Performa Model SVM (Tugas 1) ---")
print(results_df_sorted.to_markdown(index=False, floatfmt=".4f"))
```
